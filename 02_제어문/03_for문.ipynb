```python
# for문의 기본구조
# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할문장1
#     수행할문장2
#     ...

num_list = ["one", "two", "three"]
for num in num_list:
    print(num)
```

```python
a = [(1, 2), (3, 4), (5, 6)]
for i in a:
    print(i)
```

```python
tuple_list = [(1, 2), (3, 4), (5, 6)]
for a in tuple_list:
    print(a)
```

```python
a, b = (1, 2)
[a, b] = [1, 2]
[a, *b] = (1, 2, 3)
```

```python
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks:
    number = number + 1
    if mark >= 60:
        print(f"{number}번째 학생은 통과입니다.")
    else:
        print(f"{number}번째 학생은 불합격입니다.")
```

```python
# for문과 continue문
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks:
    number += 1
    if mark < 60:
        continue
    print(f"{number}번째 학생은 합격입니다.")
```

```python
result = 0
for number in range(1, 10+1):
    result = result + number # result += number
print(result)
```

```python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print(f"{number+1}번째 학생은 합격입니다.")
```

```python
# for와 range를 이용한 구구단
for i in range(2, 10):
    for j in range(1, 10):
        print(i * j, end=" ")
    print("")
```

```python
# # 출력결과를
# 2 x 1 = 2
# 2 x 2 = 4
# ...
# 2 x 9 = 18

# 3 x 1 = 3
# ...
# ...
# 9 x 9 = 81

for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i} x {j} = {i * j}")
    print("")
```

```python
# *
# **
# ***
# ****
# *****
for i in range(1, 6):
    print("*" * i)
```

```python
#     *
#    **
#   ***
#  ****
# *****
for i in range(1, 6):
    print(" " * (5 - i) + "*" * i)
```

```python
# 리스트 컨프리헨션 사용하기
a = [1, 2, 3, 4]
result = []
for num in a:
    result.append(num * 3)

print(result)
```

```python
a = [1, 2, 3, 4]
result = [num * 3 for num in a]
print(result)
```

```python
# for문과 break문
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# enumerate()함수 활용하기
fruits = ["apple", "banana", "orange"]
for i, fruit in enumerate(fruits):  # 서비스로 0부터 순차적인 값
    print(f"{i + 1}번째 : {fruit}")
```

```python
# zip함수로 여러 리스트 함께 순회하기
names = ["홍길동", "유재석", "강호동"]
scores = [60, 70, 80]
for name, score in zip(names, scores):
    print(f"{name} : {score}점")
```

```python
# 3개 이상의 리스트 - zip()
names = ["홍길동", "유재석", "강호동"]
Korean = [60, 70, 80]
English = [80, 90, 70]

for name, score1, score2 in zip(names, Korean, English):
    print(f"{name}의 국어: {score1}, 영어: {score2}")
```
