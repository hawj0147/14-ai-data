```python
# 돈이 3만원 이상이 있거나 카드가 있다면 택시를 타고가고,
# 그렇지 않으면 걸어가라

money = 30000
card = True
if money >= 30000 or card == True:
    print("택시를 타고가라")
else: 
    print("걸어가라")
```

```python
# 만약에 주머니에 돈(money)이 있으면 택시를 타고 가고,
# 없으면 걸어가라

pocket = ["paper", "phone", "money"]

if "money" in pocket:
    print("택시를 타고가라")
else:
    print("걸어가라")
```

```python
# 다양한 조건을 판단하는 elif

# 주머니에 돈이 있으면 택시를 타고 가고
# 주머니에 돈이 없지만 카드가 있으면 택시를 타고 가고
# 돈도 없고 카드도 없으면 걸어가라

pocket = ["paper", "phone", "money"]
card = True

if "money" in pocket:
    print("택시를 타고 가라")
else:
    if card == True:
        print("택시를 타고 가라")
    else:
        print("걸어가라")
```

```python
# 성적이 90점 이상이면 "A학점"
# 성적이 90점 미만이면서 80점 이상이면 "B학점"
# 성적이 80점 미만이면서 70점 이상이면 "C학점"
# 성적이 70점 미만이면서 60점 이상이면 "D학점"
# 60점 미만이면 "F학점"

score = 80

if score >= 90: print("A학점")
elif score >= 80: print("B학점")
elif score >= 70: print("C학점")
elif score >= 60: print("D학점")
else: print("F학점")
```

```python
# macth-case 구문

# match 변수
#   case 패턴1: 
#       수행할문장
#   case 패턴2:
#       수행할문장
```

```python
grade = input("A~F 중 하나의 값을 입력하시오")

match grade:
    case "A" | "B" | "C":
        print("합격입니다")
    case _: # 그 외
        print("불합격입니다")
```
