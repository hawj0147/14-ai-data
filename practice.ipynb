```python
num = int(input("숫자를 입력하세요: "))
if num <= 0:
    print("X")
else:
    i = 1
    total = 0    
    expression = ""
    while i <= num:
        total += i
        expression += str(i)
        if i != num:
            expression += " + "
        i += 1
    print(f"{expression} = {total}")
```

```python
odds =[]
evens = []

for i in range(1, 21):
    if i % 2 ==0:
        evens.append(i)
    else:
        odds.append(i)
print(f"홀수: {odds} (총{len(odds)}개)")
print(f"짝수: {evens} (총{len(evens)}개)")

", ".join(list(map(str, odds)))
```

```python
import random
number = int(random.random() * 100) + 1
i = 0
print("1부터 100사이의 숫자를 맞춰보세요!")
while True:
    i += 1
    choice = int(input(f"시도 {i}: "))
    print(f"시도 {i}: {choice}")
    if choice > number:
        print("Down!")
    elif choice < number:
        print("Up!")
    elif choice == number:
        print(f"정답입니다! {i}번 만에 맞추셨습니다!")
        break
```

```python
# 몇 번째까지는 30을 더하고 그 이후부터 끝까지 40을 더하는 함수

def calculate(a, b, c):
    total = a
    ls = []
    for i in range(b):
        print(total)
        total += 30
        ls.append(total)
    for i in range(b, c+1):
        print(total)
        total += 40
        ls.append(total)


    print(sum(ls))

calculate(30, 4, 9)
```

```python
def left_triangle(height):
    for i in range(1, height+1):
        print("*" * i)

left_triangle(5)
```

```python
def find_max(a, b, c):
    maxnum = a
    if b > maxnum:
        maxnum = b
    if c > maxnum:
        maxnum = c
    return maxnum

def max_find(a, b, c):
    numbers = [a, b, c]
    maxnum = numbers[0]
    for num in numbers:
        if num > maxnum:
            maxnum = num
    return maxnum

print(find_max(3, 8, 5))
print(max_find(3, 8, 5))
```

```python
def my_len(text):
    count = 0
    for _ in text:
        count += 1
    return count
print(my_len("python"))
```
