```python
num = int(input("숫자를 입력하세요: "))
if num <= 0:
    print("X")
else:
    i = 1
    total = 0    
    expression = ""
    while i <= num:
        total += i
        expression += str(i)
        if i != num:
            expression += " + "
        i += 1
    print(f"{expression} = {total}")
```

```python
odds =[]
evens = []

for i in range(1, 21):
    if i % 2 ==0:
        evens.append(i)
    else:
        odds.append(i)
print(f"홀수: {odds} (총{len(odds)}개)")
print(f"짝수: {evens} (총{len(evens)}개)")

", ".join(list(map(str, odds)))
```

```python
import random
number = int(random.random() * 100) + 1
i = 0
print("1부터 100사이의 숫자를 맞춰보세요!")
while True:
    i += 1
    choice = int(input(f"시도 {i}: "))
    print(f"시도 {i}: {choice}")
    if choice > number:
        print("Down!")
    elif choice < number:
        print("Up!")
    elif choice == number:
        print(f"정답입니다! {i}번 만에 맞추셨습니다!")
        break
```
