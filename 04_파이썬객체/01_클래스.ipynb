```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result = self.result + num
        return self.result
    
    def sub(self, num):
        self.result -= num
        return self.result
    
    def mul(self, num):
        self.result *= num
        return self.result
    
    def div(self, num):
        if num == 0:
            return "ERROR. 0으로 나눌 수 없습니다."
        self.result = self.result / num
        return self.result
    
cal1 = Calculator()
print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

print("==================================")

cal2 = Calculator()
print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.div(5))
print(cal2.div(0))
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result

    def div(self):
        result = self.first / self.second
        return result
    
cal1 = FourCal()
cal1.setdata(7, 2)
print(cal1.add())
print(cal1.sub())
print(cal1.mul())
print(cal1.div())
```
