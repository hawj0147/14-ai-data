```python
# 큰 따옴표 둘러싸기
"Life is too short, you need python"

# 작은 따옴표 둘러싸기
'Python is fun'

# 큰 따옴표 3개를 연속으로 둘러싸기
"""Life is too short,
 you need python"""

# 작은 따옴표 3개를 연속으로 둘러싸기
'''Life is too short,
 you need python'''
```

```python
food = "python's favorite food is perl"
say = '"Python is very easy." he says'

# 역슬래시를 사용해서 작은따옴표와 큰따옴표를 문자열에 포함하기
# 이스케이프 코드

food = 'python\'s favorite food is perl'
food

say = "\"Python is very easy.\" he says"
```

```python
#줄을 바꾸기 위한 이스케이프 코드 \n
a = "Life is too short, \n you need python"
print(a)
```
