```python
print(1, bool(1))
print(0, bool(0)) # 0만 False
print(2, bool(2))
print("python", bool("python"))
print("", bool("")) # 문자열에서는 아무것도 없는 것만 False
print("[]", bool([])) # 리스트에서는 비어있는 것만 False
print("[1, 2, 3]", bool([1, 2, 3]))
```

```python
# 논리연산자
# 1. and 연산자
print(True and True)
print(True and False)
print(False and False)
print(True and True and False)

# 2. or 연산자
print(True or True)
print(True or False)
print(False or False)
print(False or False or True)

# 3. not 연산자
print(not True)
print(not False)
print(not 1)
print(not [])
```
